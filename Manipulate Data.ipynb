```python
import pandas as pd
from datetime import date
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
month = pd.read_csv('monthly_sales.csv')
```

```python
month['Month, Year'] = pd.to_datetime(month['Month, Year'], format='%b-%y')
```

```python
month['HM'] = month['HM-NE'] + month['HM-SW'] + month['HM-NW'] + month['HM-SE'] + month['HM-C']
```

```python
month['CF'] = month['CF-NE'] + month['CF-SW'] + month['CF-NW'] + month['CF-SE'] + month['CF-C']
```

```python
month['FF'] = month['FF-NE'] + month['FF-SW'] + month['FF-NW'] + month['FF-SE'] + month['FF-C']
```

```python
month_sum = month[['Month, Year', 'HM', 'CF', 'FF']]
```

```python
year2016 = pd.Timestamp(2016, 1, 1)
year2017 = pd.Timestamp(2017, 1, 1)
year2018 = pd.Timestamp(2018, 1, 1)
year2019 = pd.Timestamp(2019, 1, 1)
year2020 = pd.Timestamp(2020, 1, 1)
```

```python
monthly_data_2016 = month_sum[(month_sum['Month, Year'] >= year2016) & (month_sum['Month, Year'] < year2017)]

monthly_data_2017 = month_sum[(month_sum['Month, Year'] >= year2017) & (month_sum['Month, Year'] < year2018)]

monthly_data_2018 = month_sum[(month_sum['Month, Year'] >= year2018) & (month_sum['Month, Year'] < year2019)]

monthly_data_2019 = month_sum[(month_sum['Month, Year'] >= year2019) & (month_sum['Month, Year'] < year2020)]
```

```python
#output each year's data
monthly_data_2016['Month, Year'] = monthly_data_2016['Month, Year'].apply(lambda x: x.strftime('%Y-%m')).copy()
monthly_data_2017['Month, Year'] = monthly_data_2017['Month, Year'].apply(lambda x: x.strftime('%Y-%m')).copy()
monthly_data_2018['Month, Year'] = monthly_data_2018['Month, Year'].apply(lambda x: x.strftime('%Y-%m')).copy()
monthly_data_2019['Month, Year'] = monthly_data_2019['Month, Year'].apply(lambda x: x.strftime('%Y-%m')).copy()
```

```python
monthly_data_2016.set_index('Month, Year', inplace = True)
monthly_data_2017.set_index('Month, Year', inplace = True)
monthly_data_2018.set_index('Month, Year', inplace = True)
monthly_data_2019.set_index('Month, Year', inplace = True)
```

```python
monthly_data_2016.to_csv('monthly_data_2016.csv')
monthly_data_2017.to_csv('monthly_data_2017.csv')
monthly_data_2018.to_csv('monthly_data_2018.csv')
monthly_data_2019.to_csv('monthly_data_2019.csv')
```
